```python
def split_image(image_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Open the image
    img = Image.open(image_path)
    img_width, img_height = img.size
    
    # Calculate the size of each sub-image
    sub_image_width = img_width // 3
    sub_image_height = img_height // 3
    
    # Loop through the grid and save each sub-image
    for row in range(3):
        for col in range(3):
            left = col * sub_image_width
            upper = row * sub_image_height
            right = left + sub_image_width
            lower = upper + sub_image_height
            
            sub_image = img.crop((left, upper, right, lower))
            sub_image.save(os.path.join(output_folder, f"deadCat_{row}_{col}.png"))

# Example usage
```

```python
from PIL import Image, ImageDraw, ImageFont
import os

def create_image_grid_with_text(image_folder, text_list, output_path, grid_size=(3, 3), image_size=(100, 100), font_path=None, font_size=20):
    """
    Arrange images in a grid with text surrounding them.

    :param image_folder: Path to the folder containing images.
    :param text_list: List of strings to display around the grid.
    :param output_path: Path to save the resulting image.
    :param grid_size: Tuple indicating the number of rows and columns in the grid.
    :param image_size: Tuple indicating the size of each image in the grid.
    :param font_path: Path to the font file (optional).
    :param font_size: Size of the font for the text.
    """
    # Load images
    image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
    images = [Image.open(os.path.join(image_folder, f)).resize(image_size) for f in image_files[:grid_size[0] * grid_size[1]]]

    # Create a blank canvas
    grid_width = grid_size[1] * image_size[0]
    grid_height = grid_size[0] * image_size[1]
    margin = 50  # Margin for text
    canvas_width = grid_width + 2 * margin
    canvas_height = grid_height + 2 * margin
    canvas = Image.new('RGB', (canvas_width, canvas_height), 'white')

    # Draw the grid
    for i, img in enumerate(images):
        row, col = divmod(i, grid_size[1])
        x = margin + col * image_size[0]
        y = margin + row * image_size[1]
        canvas.paste(img, (x, y))

    # Draw the text
    draw = ImageDraw.Draw(canvas)
    font = ImageFont.truetype(font_path, font_size) if font_path else ImageFont.load_default()
    text_margin = 10

    # Top text
    if len(text_list) > 0:
        draw.text((canvas_width // 2, text_margin), text_list[0], fill='black', font=font, anchor="mm", angle=270)

    # Bottom text
    if len(text_list) > 1:
        draw.text((canvas_width // 2, canvas_height - text_margin), text_list[1], fill='black', font=font, anchor="mm")

    # Left text
    if len(text_list) > 2:
        draw.text((text_margin, canvas_height // 2), text_list[2], fill='black', font=font, anchor="mm")

    # Right text
    if len(text_list) > 3:
        draw.text((canvas_width - text_margin, canvas_height // 2), text_list[3], fill='black', font=font, anchor="mm", angle=90)
    # Save the resulting image
    canvas.save(output_path)

# Example usage
create_image_grid_with_text("output_folder", ["Top Text", "Bottom Text", "Left Text", "Right Text"], "output_image.png")
```

```python
import os

def crop_black_border(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for filename in os.listdir(input_folder):
        if filename.endswith((".png", ".jpg", ".jpeg")):
            image_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            
            # Open the image
            img = Image.open(image_path).convert("RGB")
            # Get the bounding box of non-black pixels
            bbox = img.getbbox()
            if bbox:
                # Crop the image to the bounding box
                cropped_img = img.crop(bbox)
                # Save the cropped image
                cropped_img.save(output_path)
            else:
                print(f"The image {filename} is fully black or empty.")

# Example usage
crop_black_border("output_folder", "output_folder/cropped")
```
